```python
import pandas as pd
from sklearn import metrics
import seaborn as sns
```

```python
data = pd.read_csv('problem_2_train.csv')
```

```python
data.head(20)
```

```python
data['claimtype'].value_counts()
```

```python
data.drop(columns=['hubpincode','memcreateddate','srno','mobilepurchasedate','claimtype'],inplace=True)
```

```python
data.head()
```

```python
data['custpincodecategory'].value_counts()
```

```python
data.isnull().values.any()
```

```python
data.dtypes
```

```python
data.claimintdate = data.claimintdate.map(lambda x:x+" 00:00:00")
data.claimintdate = pd.to_datetime(data.claimintdate).astype('int64')
```

```python
data.pickupStartdate = data.pickupStartdate.map(lambda x:x+" 00:00:00")
data.pickupStartdate = pd.to_datetime(data.pickupStartdate).astype('int64')
```

```python
data.head()
```

```python
feature_cols = ['mobmake','hubid','pickuppartnerid','deliverypartnerid','insuranceCompanyId','custpincodecategory','custpincode', 'claimintdate',
'servicecntrid','pickupStartdate']
X = data[feature_cols]
y = data['tat_in_days']
```

```python
X.mobmake.value_counts()
```

```python
#using mobmake as categorical data
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder=LabelEncoder()
X['mobmake'] = labelEncoder.fit_transform(X['mobmake'])
```

```python
X.head()
```

```python
X.custpincode.value_counts()
```

```python
X.custpincodecategory.value_counts()
```

```python
X.custpincodecategory = labelEncoder.fit_transform(X.custpincodecategory)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,random_state=1,test_size=0.25)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train, y_train)
y_pred = linreg.predict(x_test)
```

```python
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_pred,y_test))
```

```python
list(zip(feature_cols, linreg.coef_))
```

# GradientBoost Regression

```python
from sklearn.ensemble import GradientBoostingRegressor
gradientBoostingRegressor = GradientBoostingRegressor()
gradientBoostingRegressor.fit(x_train,y_train)
y_pred = gradientBoostingRegressor.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```

# using KNN

```python
from sklearn.neighbors import KNeighborsRegressor
kNeighborsRegressor = KNeighborsRegressor()
kNeighborsRegressor.fit(x_train,y_train)
y_pred = kNeighborsRegressor.predict(x_test)
```

```python
np.sqrt(mean_squared_error(y_pred,y_test))
```
